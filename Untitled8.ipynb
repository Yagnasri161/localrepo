```python
import time
import numpy as np

# Define the size of the arrays/lists
size = 10**6

# Python List performance
start_time = time.time()
python_list = list(range(size))
sum_python_list = sum(python_list)
end_time = time.time()
python_list_time = end_time - start_time
print(f"Time taken by Python list: {python_list_time:.6f} seconds")

# NumPy Array performance
start_time = time.time()
numpy_array = np.arange(size)
sum_numpy_array = np.sum(numpy_array)
end_time = time.time()
numpy_array_time = end_time - start_time
print(f"Time taken by NumPy array: {numpy_array_time:.6f} seconds")

print(f"\nNumPy is {python_list_time / numpy_array_time:.2f} times faster than Python list for this operation.")
```
